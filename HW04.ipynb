````python
import numpy as np

# Step 1: Create a random sample
np.random.seed(42)  # For reproducibility
sample_size = 30 
sample = np.random.normal(loc=100, scale=15, size=sample_size)  # A sample from a normal distribution

# Step 2: Function to generate bootstrap samples and calculate the bootstrap confidence interval
def bootstrap_confidence_interval(data, num_bootstrap=1000, confidence_level=0.95):
    bootstrap_means = []
    for _ in range(num_bootstrap):
        # Create a bootstrap sample by sampling with replacement from the original data
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the mean of the bootstrap sample
        bootstrap_means.append(np.mean(bootstrap_sample)) #TO CHANGE TO OTHER PARAMETERS THIS NEEDS TO BE CHANGED TO WHATEVER YOU WANT. Change it from mean to median...
    
    # Step 3: Calculate the confidence interval
    lower_percentile = (1 - confidence_level) / 2
    upper_percentile = 1 - lower_percentile
    ci_lower = np.percentile(bootstrap_means, lower_percentile * 100)#instead of having bootstrap_means, you would for example, bootstrap_medians.
    ci_upper = np.percentile(bootstrap_means, upper_percentile * 100)# ^^^
    
    return ci_lower, ci_upper

# Step 4: Calculate the 95% confidence interval using bootstrap
bootstrap_ci = bootstrap_confidence_interval(sample)

bootstrap_ci
````

````python
import pandas as pd

# Load the data
file_path = 'https://raw.githubusercontent.com/JamesPapadakos/STA130-Work-/refs/heads/main/Untitled%20spreadsheet%20-%20Sheet1.csv'
data = pd.read_csv(file_path)

# Calculate the absolute distance between InitialHealthScore and FinalHealthScore
data['ScoreDifference'] = abs(data['InitialHealthScore'] - data['FinalHealthScore'])

# Calculate the average distance
average_distance = data['ScoreDifference'].mean()

# Print the result
print(f'The average distance between Initial and Final Health Scores is: {average_distance:.2f}')
````

This code takes all the difference in health scores and takes the mean to show on average the improvement or deterioration of the health. In this case because the value is positive, the patients final health score will improve an average of 4.50 points compared to the initial. 

````python
import numpy as np
import plotly.graph_objects as go
import pandas as pd

# Seed for reproducibility
np.random.seed(42)

# Assuming data is a pandas DataFrame with 'InitialHealthScore' and 'FinalHealthScore'
# If the data is not provided, you can replace it with an actual DataFrame

# Step 1: Calculate the difference between InitialHealthScore and FinalHealthScore
data['HealthScoreDifference'] = data['FinalHealthScore'] - data['InitialHealthScore']

# Step 2: Define a bootstrapping function
def bootstrap_resample(data, n_resamples=1000):
    resamples = []
    for _ in range(n_resamples):
        resample = np.random.choice(data, size=len(data), replace=True)
        resamples.append(np.mean(resample))
    return resamples

# Step 3: Perform bootstrapping on the HealthScoreDifference
bootstrapped_differences = bootstrap_resample(data['HealthScoreDifference'], n_resamples=1000)

# Step 4: Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrapped_differences, 2.5)
upper_bound = np.percentile(bootstrapped_differences, 97.5)

# Step 5: Create the Plotly histogram
fig = go.Figure()

# Add the histogram of bootstrapped resamples
fig.add_trace(go.Histogram(
    x=bootstrapped_differences,
    nbinsx=30,
    marker=dict(color='blue', line=dict(color='black', width=1)),
    opacity=0.7
))

# Add the 95% confidence interval as vertical lines
fig.add_vline(x=lower_bound, line=dict(color='red', dash='dash'), name='Lower 95% CI')
fig.add_vline(x=upper_bound, line=dict(color='green', dash='dash'), name='Upper 95% CI')

# Update layout for better appearance
fig.update_layout(
    title='Histogram of Bootstrapped Resamples (HealthScore Difference)',
    xaxis_title='Mean HealthScore Difference',
    yaxis_title='Frequency',
    showlegend=False,
    width=800,
    height=500,
    bargap=0.05,
)

# Display the plot with PNG renderer
fig.show(renderer="png")
````

This graph represents the bootstrapped values for the mean of the healthscore difference. It shows the mean values of 1000 resamples of the original data. It also shows the 95% confidence interval within the red and blue dashed lines by showing where the 2.5 percentile lies with the red line and where the 97.5 percentile lies with blue line.

Zero is not in the 95% confidence interval(as seen in the graph). Since there is 95% chance that the true population mean lies within the interval we can reject the null hypothesis of "no effect".
